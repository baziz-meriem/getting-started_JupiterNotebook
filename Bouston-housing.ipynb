```python
X = df['LSTAT'].values.reshape(-1,1)
y = df['MEDV'].values
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor()
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3, 20, 1)
line_y_ransac = ransac.predict(line_X.reshape(-1, 1))
```

```python
seab.set(style='darkgrid', context='notebook')
plot.figure(figsize=(12,12));
plot.scatter(X[inlier_mask], y[inlier_mask], 
            c='blue', marker='o', label='Inliers')
plot.scatter(X[outlier_mask], y[outlier_mask],
            c='red', marker='.', label='Outliers')
plot.plot(line_X, line_y_ransac, color='red')
plot.ylabel("Median value of owner-occupied homes in $1000's")
plot.legend(loc='upper left')
plot.show()
```
